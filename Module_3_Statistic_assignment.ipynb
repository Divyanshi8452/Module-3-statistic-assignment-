1. What is hypothesis testing in statistics?
--> Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It helps you determine whether there is enough evidence in a sample to support or reject a specific claim (or hypothesis) about a population parameter.

2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?
--> The null hypothesis (H₀) is a formal statement that there is no effect, no difference, or no relationship between variables. It represents the status quo or a baseline assumption that you test against.
- It's the hypothesis that the researcher typically tries to disprove.
- In statistical tests, the burden of proof is on showing that the null hypothesis is unlikely to be true.

The alternative hypothesis (H₁ or Ha) is what you propose if the null hypothesis is rejected. It suggests that there is an effect, difference, or relationship.
- It represents what you want to test or prove.
- It is accepted only when there's strong enough evidence to reject the null hypothesis.

3. Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
--> The significance level, denoted by α (alpha), is a key concept in hypothesis testing. It represents the threshold for how much risk you're willing to accept in making a Type I error — which means rejecting the null hypothesis (H₀) when it is actually true.

Role in Deciding the Outcome of a Test
When you run a hypothesis test, you calculate a p-value — the probability of getting the observed results (or more extreme) if the null hypothesis is true.

4. What are Type I and Type II errors? Give examples of each
-->In hypothesis testing, you're making a decision based on data — either to reject or not reject the null hypothesis (H₀). Since this decision is based on a sample, there's always a chance of making a mistake.

Type I Error (False Positive)
- Definition: Rejecting the null hypothesis when it is actually true.
- In simple terms: You think there's an effect or difference, but in reality, there isn't.
- Probability: This error is controlled by the significance level (α).

Example:
Imagine a medical test for a disease:
- H₀: The patient does not have the disease.
- You reject H₀ and say the patient has the disease.
- But in reality, the patient is healthy.
This is a Type I error — a false alarm.

Type II Error (False Negative)
- Definition: Failing to reject the null hypothesis when it is actually false.
- In simple terms: You miss a real effect or difference.
- Probability: Denoted by β (beta); reducing this error often requires larger sample sizes or more sensitive tests.

Example:
Using the same medical test:
- H₀: The patient does not have the disease.
- You fail to reject H₀, concluding the patient is healthy.
- But in reality, the patient does have the disease.
This is a Type II error — a missed detection.

5. What is the difference between a Z-test and a T-test? Explain when to use
each.
--> Both Z-tests and T-tests are used in hypothesis testing to determine if there's a significant difference between sample data and a population parameter (like a mean). The key difference lies in the information you have and the sample size.

When to Use Each

**Use a Z-test when**
- The population standard deviation (σ) is known
- The sample size is 30 or more
- Data is approximately normally distributed

**Use a T-test when**
- The population standard deviation is unknown
- The sample size is less than 30
- The data is approximately normal (or the sample size is large enough for normal approximation)

6. Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.
Hint: Generate random number using random function.









```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10        # number of trials
p = 0.5       # probability of success
size = 1000   # number of samples to generate

# Generate binomial distribution data
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(range(n+1))
plt.show()
```

7. : Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Known population parameters
mu_0 = 50.0       # population mean
sigma = 0.5       # population standard deviation (assumed known)

# Sample statistics
x_bar = np.mean(sample_data)
n = len(sample_data)

# Z-statistic
z = (x_bar - mu_0) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Print results
print(f"Sample Mean (x̄): {x_bar:.4f}")
print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H₀): There is a significant difference.")
else:
    print("Fail to reject the null hypothesis (H₀): No significant difference.")
```

8. Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data
np.random.seed(42)  # for reproducibility
mu = 100            # true mean
sigma = 15          # true standard deviation
n = 100             # sample size

data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
se = sample_std / np.sqrt(n)

# Step 3: Calculate 95% confidence interval
confidence = 0.95
z_score = stats.norm.ppf(1 - (1 - confidence)/2)
margin_of_error = z_score * se

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 4: Plot histogram and confidence interval
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='dotted', linewidth=2, label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='dotted', linewidth=2, label=f"95% CI Upper = {ci_upper:.2f}")

plt.title('Histogram of Simulated Normal Data\nwith 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

-->
Explanation: What Do Z-Scores Represent?

- A Z-score tells you how many standard deviations a data point is from the mean.
If a Z-score is:
- 0, the value is exactly at the mean.
- Positive, the value is above the mean.
- Negative, the value is below the mean.

For example, a Z-score of +2 means the value is 2 standard deviations above the mean.
Z-scores help standardize data for comparison, especially when datasets have different scales.

```python

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores from the data and plot their histogram.

    Parameters:
    data (list or np.array): Input dataset

    Returns:
    z_scores (np.array): Array of Z-scores
    """
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot histogram of Z-scores
    plt.hist(z_scores, bins=15, edgecolor='black', color='skyblue', alpha=0.7)
    plt.title('Histogram of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores

# Example usage
data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5]

z_scores = calculate_and_plot_z_scores(np.array(data))
print("First 5 Z-scores:", z_scores[:5])
```
